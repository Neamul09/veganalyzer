```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure visuals
%matplotlib inline
sns.set(style="whitegrid", palette="muted")
```

```python
# Step 2: Load your CSV file
df = pd.read_csv("final_data.csv", parse_dates=["date"])

# Preview the data
df.head(100)
```

```python
# Step 3: Check for missing values and basic info
df.info()
df.describe()
df.isnull().sum()
```

```python
# Step 4: Plot price and trend scores
plt.figure(figsize=(14, 6))
plt.plot(df["date"], df["avg"], label="Avg Price", linewidth=2)
plt.plot(df["date"], df["trend_score_পেঁয়াজ"], label="Trend পেঁয়াজ", linestyle="--")
plt.plot(df["date"], df["trend_score_onion"], label="Trend Onion", linestyle="--")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Price vs. Google Trends")
plt.legend()
plt.show()
```

```python
# Step 5: Check correlation between lagged trends and price
lag_columns = ["trend_score_পেঁয়াজ_lag1", "trend_score_onion_lag1"]
df[lag_columns + ["avg"]].corr()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(df["date"], df["avg"], label="Avg Price")
plt.plot(df["date"], df["trend_score_পেঁয়াজ_lag1"], label="Trend পেঁয়াজ Lag1")
plt.plot(df["date"], df["trend_score_onion_lag1"], label="Trend Onion Lag1")
plt.xlabel("Date")
plt.title("Lead-Lag Relationship")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(df["date"], df["avg"], label="Avg Price", color="gray")
plt.scatter(df[df["trend_score_পেঁয়াজ_spike"] == 1]["date"],
            df[df["trend_score_পেঁয়াজ_spike"] == 1]["avg"],
            label="পেঁয়াজ Spike", color="red", s=100, marker='^')
plt.scatter(df[df["trend_score_onion_spike"] == 1]["date"],
            df[df["trend_score_onion_spike"] == 1]["avg"],
            label="Onion Spike", color="blue", s=100, marker='v')
plt.title("Price with Trend Spikes")
plt.xlabel("Date")
plt.ylabel("Avg Price")
plt.legend()
plt.show()
```

```python
# Optional: Correlation heatmap for all trend vs price columns
trend_cols = [col for col in df.columns if "trend_score" in col] + ["avg"]
plt.figure(figsize=(12, 8))
sns.heatmap(df[trend_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Weather vs Price

plt.figure(figsize=(16, 10))

plt.subplot(3, 1, 1)
plt.plot(df["date"], df["avg"], label="Avg Price", color="black")
plt.plot(df["date"], df["temp"], label="Temperature", color="red")
plt.legend()
plt.title("Price and Temperature Over Time")

plt.subplot(3, 1, 2)
plt.plot(df["date"], df["avg"], label="Avg Price", color="black")
plt.plot(df["date"], df["rainfall"], label="Rainfall", color="blue")
plt.legend()
plt.title("Price and Rainfall Over Time")

plt.subplot(3, 1, 3)
plt.plot(df["date"], df["avg"], label="Avg Price", color="black")
plt.plot(df["date"], df["humidity"], label="Humidity", color="green")
plt.legend()
plt.title("Price and Humidity Over Time")

plt.tight_layout()
plt.show()
```

```python
#correlation with weather, trends and price

cols = ["avg", "temp", "rainfall", "humidity"] + [c for c in df.columns if "trend_score" in c]
plt.figure(figsize=(12, 8))
sns.heatmap(df[cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation: Price, Weather & Trends")
plt.show()
```

```python
# Create a condition for price surge (example: price > avg + std)
price_threshold = df["avg"].mean() + df["avg"].std()
df["price_surge"] = df["avg"] > price_threshold

# Check when trend spikes and rainfall spikes coincide with price surge
df["rain_spike"] = df["rainfall"] > df["rainfall"].mean() + df["rainfall"].std()

plt.figure(figsize=(12, 6))
plt.plot(df["date"], df["avg"], label="Avg Price", color="gray")

plt.scatter(df[(df["trend_score_পেঁয়াজ_spike"] == 1) & (df["price_surge"])].date,
            df[(df["trend_score_পেঁয়াজ_spike"] == 1) & (df["price_surge"])]["avg"],
            color="red", label="পেঁয়াজ Trend Spike + Price Surge", s=100, marker="^")

plt.scatter(df[(df["rain_spike"]) & (df["price_surge"])].date,
            df[(df["rain_spike"]) & (df["price_surge"])]["avg"],
            color="blue", label="Rainfall Spike + Price Surge", s=100, marker="o")

plt.legend()
plt.title("Price Surges with Trend and Rainfall Spikes")
plt.show()
```
